```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams.update({"font.size": 17, 'figure.figsize': [15, 7]})
```

```python
results = pd.read_csv("brute_force_agent__evaluation_experiment.csv", index_col=0)
results.head()
```

```python
plt.figure()
min_idx = np.argmin(results["utility"].values)
plt.scatter(results["w"], results["utility"])
plt.scatter(results["w"].iloc[min_idx], results["utility"].iloc[min_idx], color="red")
plt.xlabel("w")
plt.ylabel("utility")
plt.title(f"Utility min={results['utility'].iloc[min_idx]}, @ w={results['w'].iloc[min_idx]}")
plt.show()
```

```python
plt.figure()
min_idx = np.argmin(results["utility"].values)
plt.scatter(results["max_search_depth"], results["utility"])
plt.scatter(results["max_search_depth"].iloc[min_idx], results["utility"].iloc[min_idx], color="red")
plt.xlabel("max_search_depth")
plt.ylabel("utility")
plt.title(f"Utility min={results['utility'].iloc[min_idx]}, @ max_search_depth={results['max_search_depth'].iloc[min_idx]}")
plt.show()
```

```python
plt.figure()
min_idx = np.argmin(results["utility"].values)
plt.scatter(results["default_action"], results["utility"])
plt.scatter(results["default_action"].iloc[min_idx], results["utility"].iloc[min_idx], color="red")
plt.xlabel("default_action")
plt.ylabel("utility")
plt.title(f"Utility min={results['utility'].iloc[min_idx]}, @ default_action={results['default_action'].iloc[min_idx]}")
plt.show()
```

```python
plt.figure()
plt.hist(results["utility"].loc[results["max_search_depth"]==3], label="max_search_depth=3")
plt.hist(results["utility"].loc[results["max_search_depth"]==4], label="max_search_depth=4")
plt.hist(results["utility"].loc[results["max_search_depth"]==5], label="max_search_depth=5")
plt.xlabel("utility")
plt.ylabel("count")
plt.legend()
plt.show()
```
